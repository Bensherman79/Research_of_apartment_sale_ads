# Исследование объявлений о продаже квартир

## Описание проекта
В вашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Вам нужно научиться определять рыночную стоимость объектов недвижимости. Для этого проведите исследовательский анализ данных и установите параметры, влияющие на цену объектов. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 
По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта и других объектов — эти данные автоматически получены из геосервисов. Количество парков и водоёмов также заполняется без участия пользователя.

### Описание данных
- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах (м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — общая площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

```python
# заменяем пропуски в столбце is_apartment на False
data['is_apartment'] = data['is_apartment'].fillna(False)
# проверим количество пропусков
data['is_apartment'].isna().sum()
```

```python
# Заполняем пропуски по высоте потолков медианным значением относительно населенных пунктов
median_height = data.groupby('locality_name')['ceiling_height'].transform('median')
data['ceiling_height'] = data['ceiling_height'].fillna(median_height)
```

```python
# посмотрим теперь сколько значений в столбце ceiling_height
data['ceiling_height'].count()
```

```python
# Проверим количество оставшихся пропусков
data['ceiling_height'].isna().sum()
```

## Добавьте в таблицу новые столбцы

Произведём новые расчёты и добавим их в нашу таблицу для дальнейшего исследования:
- price_of_one_square_meter: цена одного квадратного метра;
- day_week_published: день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
- month_week_published: месяц публикации объявления;
- year_week_published: год публикации объявления;
- floor_category: тип этажа квартиры (значения — «первый», «последний», «другой»);
- city_centers_nearest_km: расстояние до центра города в километрах (переведите из м в км и округлите до ближайших целых значений).

```python
# найдём цену одного квадратного метра и добавим данный столбец в нашу таблицу
data['price_of_one_square_meter'] = data['last_price'] / data['total_area']
# округлим цену за один уквадрвтный метр до двух знаков после запятой
data['price_of_one_square_meter'] = data['price_of_one_square_meter'].round(2)
```

```python
# постотрим на первые 5 строк нашей таблицы
data.head()
```

```python
# добавим столбец с днём недели публикации объявления
data['day_week_published'] = data['first_day_exposition'].dt.weekday
# добавим столбец с месяцем публикации объявления
data['month_week_published'] = data['first_day_exposition'].dt.month
# добавим столбец с годом публикации объявления
data['year_week_published'] = data['first_day_exposition'].dt.year
```

```python
# постотрим на первые 5 строк нашей таблицы
data.head()
```

```python
# добавим столбец с категоризацией по этажам floor_category, а именно функцию
def get_floor_category(row):
    floor = row['floor']
    floors_total = row ['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
```

```python
# добавляем столбец с категорией этажа в нашу таблицу
data['floor_category'] = data.apply(get_floor_category, axis=1)
```

```python
# постотрим на первые 5 строк нашей таблицы
data.head()
```

```python
# добавим столбец с расстоянием до центра города в километрах
data['city_centers_nearest_km'] = data['city_centers_nearest'] / 1000
# округлим расстояние в километрах до целого значения
data['city_centers_nearest_km'] = data['city_centers_nearest_km'].round()
```

Также необходимо учитывать, что данные об количестве дней размещения объявления (с момента публикации до момента снятия) не может нести точную информацию по следующим причинам:

- пользователь не подтвердил своевременно актуальность информации и публикация была удалена автоматически (один из вариантов такого большого количества снятий объявлений в срок 5 дней и менее).
- пользователь решил не корректировать уже существующее объявление, а решил его "перезалить".

Также необходимо обратить внимание, что по числовому описанию данных среднее значение почти 180 дней и не входит в основную долю продажи объектов недвижимости. Медиана равна 95 дням. При этом среднее отклонение больше среднего значения и составляет чуть больше 218 дней. Здесь явно большое количество выбросов. Необходимо построить диаграмму размаха. Исходя из данных первой гистограммы укажем по оси ординат (y) количество дней размещения объявления от 0 до 800 (чуть больше чем половина максимального значения).

```python
# строим диаграмму размаха столбца days_exposition
data.boxplot(column='days_exposition', figsize=(10, 8))
plt.title('Размах количества дней размещения объявления')
plt.ylim(0, 800)
plt.ylabel('Количество дней')
plt.show()
```

Исходя из числового описания данных, что 75% (третий квартиль) равен 229 дней, можно легко найти по формуле где заканчивается верхняя граница "усов". Все значения, что будут выше - это выбросы и их необходимо будет удалить.
Для этого нам необходимо будет указать, что такое межквартильный размах (IQR).

P.S.: отдельное спасибо Аслану, что показал на вебинаре как легко и просто самим искать квартили и межквартильный размах)
Кстати про изучение численных параметров, их можно было также легче сделать через формулу межквартильного размаха, но я видео вебинара к сожалению посмотрел уже только после того как сделал эту часть по 10 параметрам. )))

```python
# вводим значение межквартильного размаха
q1 = data['days_exposition'].quantile(.25)
q3 = data['days_exposition'].quantile(.75)
iqr = q3 - q1
```

```python
# ищем максимальное значение межквартильного размаха от третьего квартиля (Q3)
print(q3 + 1.5 * iqr)
```

```python
# как ранее говорил, все значения, что выше в данном случае 527.5 дней - это выбросы. Удаляем их
df = data.query('days_exposition < 505')
```

```python
# проверим сколько строк были выбросами
len(data) - len(df)
```

Вывод: Убрали все выбросы из нашего датафрейма. Установили, что быстрая продажа - до 45 дней (первый квартиль (q1)), средняя продажа - 229 дней (третий квартиль (q3)), долгая продажа - до 505 дней.
